```python
import numpy as np
from tensorflow import keras
```

```python
x = np.linspace(-4,4,9)
y = np.linspace(-9,7,9)
print(x,x.shape)
print(y,y.shape)
```

```python
model = keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
model.compile(optimizer='sgd', loss='mean_squared_error')
```

```python
model.fit(x,y,epochs=600)
```

```python
print(model.predict([6, 7, 8]))
```

```python
del model
```
