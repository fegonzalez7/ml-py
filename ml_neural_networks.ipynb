### Aprendizaje profundo

```python
import tensorflow as tf
from tensorflow import keras
import numpy as np
import matplotlib.pyplot as plt
```

```python
fashion_mnist = keras.datasets.fashion_mnist
(train_im, train_labels), (test_im, test_labels) = fashion_mnist.load_data()
```

```python
class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat', 'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']
```

```python
train_im.shape # 60k imagenes
```

```python
plt.figure()
plt.imshow(train_im[10001])
plt.grid()
```

Normalizar datos

```python
train_im = train_im/255
test_im = test_im/255
```

```python
plt.figure(figsize = (10, 10))
for i in range(25):
  plt.subplot(5, 5, i + 1)
  plt.xticks([])
  plt.yticks([])
  plt.grid('off')
  plt.imshow(train_im[i], cmap = plt.cm.binary)
  plt.xlabel(class_names[train_labels[i]],color='white')
```
